```python
import json

file = open("text.txt", "r")
data = file.read()
dict = json.loads(data)
```

```python
dict
```

```python
arr_ir = []
arr_r = []

for item in dict:
    arr_ir.append(dict[item]["ir"])
    arr_r.append(dict[item]["r"])
```

```python
arr_ir, arr_r
```

```python
arr_r
```

```python

import datetime
timestamps = [datetime.datetime(2020, 5, 17,12,0,0), datetime.datetime(2020, 5, 17,12,0,1), datetime.datetime(2020, 5, 17,12,0,2)]
import matplotlib.pyplot


matplotlib.pyplot.plot_date(timestamps, arr_ir)
```

```python
import numpy as np

x = np.array(arr_ir)
y = np.array(timestamps)
matplotlib.pyplot.plot_date(y, x)
```
